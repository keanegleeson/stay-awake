```python
# This one wiggles and can be used to refresh Brave and scroll to earn a few nickles while I'm away
import pyautogui
import time
import sys
from datetime import datetime
pyautogui.FAILSAFE = False
numMin = 50
#not sure what this sys.argv stuff is doing, was able to get this to work for my purposes with just the below code
# if ((len(sys.argv)<2) or sys.argv[1].isalpha() or int(sys.argv[1])<1):
#     numMin = 3
# else:
#     numMin = int(sys.argv[1])

#this function will scroll through the news feed to hopefully catch some ads and earn $
def hold_down (hold_time):
    import time, pyautogui
    start = time.time()
    while time.time() - start < hold_time:
        pyautogui.press('down')
        
while(True):
    x=0
    while(x<numMin):
        print("sleeping")
        time.sleep(15)
        print("done sleeping")
        x+=1
        for i in range(0,200):
            pyautogui.moveTo(0,i*4)#wiggling mouse
        pyautogui.moveTo(1,1) #resets mouse
#         pyautogui.press('f5') #refreshing brave
        print("showtime")
        pyautogui.hotkey('ctrl','t') #open new tab
        pyautogui.press('tab') #get out of searchbar
        hold_down(60) #scrolling for a minute
        pyautogui.hotkey('ctrl','w') #close new tab - this might fix the scroll stopping issue
        for i in range(0,3):
            pyautogui.press("shift")
        print("Movement made at {}".format(datetime.now().time())) #noting movements
```
